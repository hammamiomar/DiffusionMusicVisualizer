```python
from imagegen import NoiseVisualizer
import numpy as np
from utils import create_mp4_from_pil_images
import os
import torch
```

```python
song="songs/Dime.mp3"
output_path="outputs/Dime.mp4"
```

```python
visualizer = NoiseVisualizer()
visualizer.loadSong(song, hop_length=377)
```

```python
latents = visualizer.getSpecCircleLatents(distance=6)
fps = visualizer.getFPS()
```

```python
latents.shape
```

```python
basePrompt = "demon dancing cheerfully with huge teeth, ugly, angry bells, creepy and morose, hardcore graphics"
targetPromptChromaScale = ["burger", "Dark Blue", "Purple", "Magenta", "happy", "Orange", "Yellow", "Sea green", 
                           "Green", "what the hell man", "Light Blue", "horse"]
alpha = 0.8
promptEmbeds = visualizer.getPromptEmbeds(basePrompt, targetPromptChromaScale, alpha=alpha)
```

```python
promptEmbeds[0].shape
```

```python
images = visualizer.getVisuals(latents=latents, promptEmbeds=promptEmbeds)
```

```python
create_mp4_from_pil_images(image_array=images,output_path=output_path,song=song, fps=fps)
```
